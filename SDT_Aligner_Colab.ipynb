# 🎙️ SDT — Subtitle from Transcript Generator

Generate accurate `.srt` subtitle files by aligning a **plain text transcript** to a local audio file using the [SDT GitHub repository](https://github.com/l2yao/SDT).

```python
# 1. Clone the repository and install dependencies
!git clone https://github.com/l2yao/SDT.git
!pip install -q stable-ts yt-dlp torch

import sys
import os
sys.path.append('/content/SDT')
from sdt.aligner import align_transcript
from sdt.srt_writer import write_srt
from google.colab import files
```

```python
# 2. Upload your MP3 and TXT files
print("Please upload your .mp3 and .txt files:")
uploaded = files.upload()
```

```python
# 3. Run Alignment (Change filenames here if needed)
AUDIO_FILE = "1401001.mp3"     # UPDATE THIS
TRANSCRIPT_FILE = "140102.txt" # UPDATE THIS
OUTPUT_SRT = "output.srt"

with open(TRANSCRIPT_FILE, "r", encoding="utf-8") as f:
    transcript_text = f.read()

print(f"🔄 Aligning {AUDIO_FILE} with {TRANSCRIPT_FILE}...")
result = align_transcript(
    audio_path=AUDIO_FILE,
    transcript=transcript_text,
    method="hybrid",
    language="zh" # Ensure this is correct for your transcript
)

write_srt(result, OUTPUT_SRT)
print(f"✅ SRT file saved to {OUTPUT_SRT}")
files.download(OUTPUT_SRT)
```
